Follow the instructions to recreate the plots using this data:

## Data

```python
import numpy as np
x = np.arange(0,100, 0.1)
y = np.cos(0.2*x)
z = np.cos(0.2*x)
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. Do you remember what command you can use when you are not working with the jupyter notebook?**

```python
from matplotlib import pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

## Exercise 1

- Create a figure object called fig.
- Add an axis to the figure at [0,0,1,1] and call this new axis ax.
- Plot x vs. y on it and set the labels and titles as shown in the plot below:

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('cos(X)')
```

## Exercise 2
- Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.5,0.6,.3,.2] respectively.

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.5,0.6,.3,.2])
```

**Let plot x vs. y on both axes. Call your figure object at the end to show it.**

```python
ax1.plot(x,y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

fig # Show figure object
```

## Exercise 3

** Now create a new plot by adding two axes to a figure located in [0,0,1,1] and [0.3,0.4,.5,.4] respectively**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.3,0.4,.5,.4])
```

**Use x,y, and z to replicate the plot below with the same x limits and y limits:**

```python
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax2.plot(x,z)
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_title('ZOOM')
ax2.set_xlim(40,55)
ax2.set_ylim(0,1)

fig
```

## Exercise 4

** Create a subplot with two rows and one column.**

```python
fig,axes= plt.subplots(nrows= 2, ncols=1)
```

**Now plot (x,z) and (x,y) on the empty axes. Let try some specific linewidths and styles**

```python
axes[0].plot(x,z,color="orange", lw=1, ls=':')
axes[1].plot(x,y,color="green", lw=3, ls='-.')

fig
```

**Can resize the plot by adding the figsize() argument in plt.subplots()?**

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(12,2))

axes[0].plot(x,z,color="blue", lw=2, ls='--')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Z')

axes[1].plot(x,y,color="yellow", lw=5)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```
